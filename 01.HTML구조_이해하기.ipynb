```python
res = urllib.request.urlopen('https://www.python.org/')
#res의 헤더에서 읽은 정보를 리스트로 가공해서 보여줌
encoding = res.info().get_content_charset()
print(res.info())
print('encoding=',encoding,'\n')
#read에 숫자를 넣으면 잘라서 보여줌
text=res.read(500).decode(encoding)
print(text)
```

```python
res = urllib.request.urlopen('https://www.daum.net/')
encoding = res.info().get_content_charset()
encoding='utf-8' if encoding == None else encoding
print(res.info())
print('encoding=',encoding,'\n')
text=res.read(500).decode(encoding)
print(text)
```

```python
#검색조건을 url에 붙이는 방법
import urllib.request
import urllib.parse

params = urllib.parse.urlencode({'name':'홍길동', 'age':20})
url = "http://unico2013.dothome.co.kr/crawling/get.php?%s" % params

res = urllib.request.urlopen(url)
#HTTPRespose가 있으므로 제대로 받은것
print(res)
```

```python
print(res.read().decode('utf-8'))
```

```python
#post방식
params = urllib.parse.urlencode({'name':'홍길동', 'age':20})
url = "http://unico2013.dothome.co.kr/crawling/post.php"
params=params.encode('ascii')
res = urllib.request.urlopen(url,params)
print(res.read().decode('utf-8'))
```

```python
#Requests 라이브러리를 이용한 호출
import requests

url = "http://unico2013.dothome.co.kr/crawling/exam.html"

res = requests.request('get',url)
res.encoding = 'utf-8'
res.text
```

```python
import requests

url = "http://unico2013.dothome.co.kr/crawling/exam.html"

res = requests.request('post',url)
res.encoding = 'utf-8'
res.text
```

```python
import requests

url_get = "http://unico2013.dothome.co.kr/crawling/get.php"
params={'name':'홍길동','age':20}
res = requests.request('get',url_get,params=params)
res.encoding = 'utf-8'
print(res.url)
print(res.text)
```

```python
import requests

url_post = "http://unico2013.dothome.co.kr/crawling/post.php"
params={'name':'강길동','age':20}
res = requests.request('post',url_post,params=params)
res.encoding = 'utf-8'
print(res.url)
print(res.text)
```
